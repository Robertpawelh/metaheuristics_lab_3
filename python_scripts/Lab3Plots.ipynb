```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib.ticker import FormatStrFormatter

sns.set_theme()
save_fig = False
```

```python
def load_optims(zad_id):
    optimums = pd.read_csv(f'../Results/zad_{zad_id}_optims.csv', sep=';')
    return optimums
```

## LAB 8

```python
def plot_metrics(zad_id, problem_names, model_names, metrics):
    markers = ['o', 's', 'v', 'D', 'P']
    for metric in metrics:
        fig, ax = plt.subplots(1, len(problem_names), figsize=(18, 4))
        for index, (problem_k, problem_name) in enumerate(problem_names.items()):
            df = pd.read_csv(f'../Results/zad_{zad_id}_{problem_k}.csv', sep=';')
            groupby = df.groupby(['method'])
            data_to_plot = []
            models = []
            for gr_id, (group_name, group) in enumerate(groupby):
                param_groups = group.sort_values(['genes']).groupby('genes')
                x = list(param_groups.genes.groups.keys())#.tolist()
                y = param_groups[metric].median().tolist()

                model_name = model_names[group_name]
                data_to_plot.append(param_groups[metric].apply(list).iloc[0])
                models.append(model_name)
            
            ax[index].boxplot(data_to_plot, medianprops = dict(linewidth=6))
            ax[index].set_title(problem_name)
            #ax[index].legend()
            ax[index].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
            # print(models)
            ax[index].set_xticks(np.arange(1, len(models) + 1, 1), models)
            ax[index].set_xlabel('Metoda')

        fig.tight_layout()
        if save_fig:
            plt.savefig(f'../Plots/zad_{zad_id}_{problem_k}_{metric}.jpg')
        else:
            print(metric)
            plt.show()
```

```python
problem_names = {'ZeroMaxOneMax': 'ZeroMaxOneMax', 
                 'Trap5InvTrap5': 'Trap5InvTrap5',
                 'MOMaxCut': 'MOMaxCut',
                 'MOKnapsack': 'MOKnapsack',
                 'LOTZ': 'LOTZ'
                }


model_names = {
    'default': 'default',
    'skrajni1': 'skrajni1',
    'skrajni2': 'skrajni2',
    'skrajni1_turniej': 'skrajni1_turniej',
    'skrajni2_turniej': 'skrajni2_turniej'
}

metrics = ['hyperVolume', 'IGD', 'lastUpdateTime', 'lastUpdateIteration', 'LastUpdateFFE']

plot_metrics(1, problem_names, model_names, metrics)
```
